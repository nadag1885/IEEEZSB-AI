# Plotting with Matplotlib
Use Matplotlib to create bar charts that visualize the conclusions you made with groupby and query.

```python
# Import necessary packages and load `winequality_edited.csv`
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# % matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
wine_df = pd.read_csv('winequality_edited.csv')
wine_df.head()
```

### #1: Do wines with higher alcoholic content receive better ratings?
Create a bar chart with one bar for low alcohol and one bar for high alcohol wine samples.

```python
# Use query to select each group and get its mean quality
low_alcohol=wine_df.query('alcohol < alcohol.median()').mean()['quality']
low_alcohol
```

```python
high_alcohol=wine_df.query('alcohol > alcohol.median()').mean()['quality']
high_alcohol
```

```python
# Create a bar chart with proper labels
plt.bar(1,high_alcohol,0.2,color='w', alpha=0.9, label='high_alcohol')
plt.bar(0.2,low_alcohol,0.2,color='r', alpha=0.9,label='low_alcohol')

plt.xlabel(' alcohol')
plt.ylabel('Quality')
plt.title('Proportion by Alcohol and Quality')
plt.legend()
```

### #2: What level of acidity receives the highest average rating?   >>>   low
Create a bar chart with a bar for each of the four acidity levels.

```python
# Use groupby to get the mean quality for each acidity level
mean=wine_df.groupby('acidity_levels')['quality'].mean()
mean
```

```python
# Create a bar chart with proper labels
ind = np.arange(len(mean))
width = 0.35  
```

```python
 plt.bar(5, mean['high'], width, color='r', alpha=.7,label='high')
 plt.bar(6, mean['low'], width, color='w', alpha=.7,label='low')  
 plt.bar(7, mean['medium'], width, color='b', alpha=.7,label='medium')
 plt.bar(8, mean['mod_high'], width, color='g', alpha=.7,label='mod_high') 
 plt.xlabel(' Acidity Levels')
 plt.ylabel('Quality')
 plt.title('Proportion by Acidity Levels and Quality')
 plt.legend()
```
